## Task 3

към data pipeline-a от книгата да се добави обработка на липсващи данни

```python
import pandas as pd

housing = pd.read_csv('datasets/housing_01.csv')
X_mean = housing.mean(axis=0, numeric_only=True).values
X_median = housing.median(axis=0, numeric_only=True).values
X_std = housing.std(axis=0, numeric_only=True).values
```

```python
n_num_inputs = 9

def parse_csv_line(line, record_defaults=None, strategy="zero"):
    if record_defaults is not None:
        defs = record_defaults
    else:
        match strategy:
            case 'zero':
                defs = [0.] * n_num_inputs + [tf.constant([], dtype=tf.string)]

            case 'mean':
                defs = X_mean.tolist() + [tf.constant([], dtype=tf.string)]

            case 'median':
                defs = X_median.tolist() + [tf.constant([], dtype=tf.string)]

    fields = tf.io.decode_csv(line, record_defaults=defs)
    return tf.stack(fields[:-1]), tf.stack(fields[-1:])

def preprocess(line, record_defaults=None, strategy="zero"):
    x, y = parse_csv_line(line, record_defaults, strategy)
    return (x - X_mean) / X_std, y


def csv_reader_dataset(filepaths, n_readers=5, n_read_threads=None,
                        n_parse_threads=5, shuffle_buffer_size=10_000, seed=42,
                        batch_size=32, record_defaults=None, strategy="zero"):
    
    dataset = tf.data.Dataset.list_files(filepaths, seed=seed)
    dataset = dataset.interleave(
        lambda filepath: tf.data.TextLineDataset(filepath).skip(1),
        cycle_length=n_readers, num_parallel_calls=n_read_threads)
    
    dataset = dataset.map(lambda line: preprocess(line, record_defaults, strategy), 
                          num_parallel_calls=n_parse_threads)
    
    dataset = dataset.shuffle(shuffle_buffer_size, seed=seed)
    return dataset.batch(batch_size).prefetch(1)
```

```python
ds = csv_reader_dataset('datasets/housing_*.csv', batch_size=3)

i = 5
for item, _ in zip(ds, range(i)):
    print(item)
```

## Task 2

да се имплеметира класа Normalization така че да работят примерите в книгата

```python
class Normalization(tf.keras.layers.Layer):
    def adapt(self, X):
        self.mean_ = tf.reduce_mean(X, axis=0, keepdims=True)
        self.std_  = tf.math.reduce_std(X, axis=0, keepdims=True)

    def call(self, inputs):
        eps = tf.keras.backend.epsilon()
        return (inputs - self.mean_) / (self.std_ + eps)
```

```python
np.random.seed(13)
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

X, y = fetch_california_housing(return_X_y=True)
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, random_state=42, test_size=0.2, shuffle=True) 
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, random_state=42, test_size=0.1, shuffle=True) 
```

```python
norm_layer = tf.keras.layers.Normalization()

model = tf.keras.models.Sequential([
    norm_layer,
    tf.keras.layers.Dense(1)
])

model.compile(loss="mse", optimizer=tf.keras.optimizers.SGD(learning_rate=2e-3))
norm_layer.adapt(X_train) 
model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=5)
```

```python
norm_layer = tf.keras.layers.Normalization()
norm_layer.adapt(X_train)
X_train_scaled = norm_layer(X_train)
X_valid_scaled = norm_layer(X_valid)

model = tf.keras.models.Sequential([tf.keras.layers.Dense(1)])
model.compile(loss="mse", optimizer=tf.keras.optimizers.SGD(learning_rate=2e-3))
model.fit(X_train_scaled, y_train, epochs=5, validation_data=(X_valid_scaled, y_valid))
```
