```python
import tensorflow as tf
import time
```

```python
@tf.function
def add_optimized(a, b):
  return a + b

def add(a, b):
  return a + b
```

```python
start = time.time()
add(tf.ones([2, 2]), tf.ones([2, 2]))  #  [[2., 2.], [2., 2.]]
normal_time = time.time() - start


start = time.time()
add_optimized(tf.ones([2, 2]), tf.ones([2, 2]))  #  [[2., 2.], [2., 2.]]
optimized_time = time.time() - start

f'normal: {normal_time}; optimized: {optimized_time}'
```
