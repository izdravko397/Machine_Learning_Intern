#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

````python
P0 = np.array([[0, 0], [5, 5]])
P1 = np.array([[1, 2], [6, 6]])
p = np.array([2, 2])

v = P1 - P0
w = p - P0

t = (w * v) / (v ** 2).sum(axis=1)

closest = P0 + t * v
dist = np.linalg.norm(p - closest, ord=2, axis=1)
dist
````

#### 79. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])? (★★★)

````python
P0 = np.array([[0, 0], [5, 5]])
P1 = np.array([[1, 2], [6, 6]])
P = np.array([[2, 2], [3, 3], [0, 1]])

res = np.empty((len(P), len(P0)))
for i in range(len(P)):
    v = P1 - P0
    w = P[i] - P0

    t = (w * v) / (v ** 2).sum(axis=1)

    closest = P0 + t * v
    res[i] = np.linalg.norm(P[i] - closest, ord=2, axis=1)

res
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
class Mat:
    def __init__(self, data):
        if not isinstance(data, np.ndarray):
            data = np.asarray(data)

        self.data = data

    def __setitem__(self, key, val):
        self.data[key] = val
        
        if isinstance(key, tuple):
            self.data[key[1], key[0]] = val

    def __str__(self):
        return str(self.data)


m = Mat(np.zeros((5, 5)))
print(m)

m[2, 3] = 1
print(m)
````

#### 100. Compute bootstrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means). (★★★)

````python
X = np.array([1, 2, 3, 4, 5])
N = 100  

boot_means = np.empty(N)
for i in range(N):
    sample = np.random.choice(X, size=len(X), replace=True)
    boot_means[i] = np.mean(sample)

lower = np.percentile(boot_means, 2.5)
upper = np.percentile(boot_means, 97.5)
lower, upper 
````
