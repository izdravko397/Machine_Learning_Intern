```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False) 

X, y = mnist.data, mnist.target
X_train, y_train = X[:10000], y[:10000]
```

```python
from sklearn.linear_model import SGDClassifier
from importlib import reload
import matplotlib.pyplot as plt
sgd_clf = SGDClassifier(random_state=42, n_jobs=1)
```

```python
def plot_digit(image_data):
    image = image_data.reshape(28, 28)
    plt.imshow(image, cmap="binary")
    plt.axis("off")
```

```python
import multi_classifiers
reload(multi_classifiers)
from multi_classifiers import OneVsRestClassifier
my_OvR_clf = OneVsRestClassifier(sgd_clf)
my_OvR_clf.fit(X_train, y_train)

from sklearn.multiclass import OneVsRestClassifier
OvR_clf = OneVsRestClassifier(sgd_clf)
OvR_clf.fit(X_train, y_train)
```

```python
print(my_OvR_clf.estimators_)
print(OvR_clf.estimators_)

print(my_OvR_clf.classes_)
print(OvR_clf.classes_)
```

```python
digit = X[50000]

plot_digit(digit)
```

```python
print(my_OvR_clf.predict([digit]))
print(OvR_clf.predict([digit]))

print(my_OvR_clf.decision_function([digit]))
print(OvR_clf.decision_function([digit]))
```

```python
print(my_OvR_clf.estimators_)
print(OvR_clf.estimators_)

print(my_OvR_clf.classes_)
print(OvR_clf.classes_)
```
